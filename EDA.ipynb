```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bees = pd.read_csv('/Users/phoebewu/Desktop/final/honey-bee-annotated-images/bee_data.csv')
bees.head()
```

```python
#check null value
bees.isnull().sum() 
```

```python
#check duplicates
duplicates = bees.duplicated()
duplicates[duplicates > 1]
duplicates = bees.file.value_counts()
duplicates[duplicates > 1]
```

```python
plt.figure(figsize=(8,8))
bees.date.value_counts().plot(kind = 'bar')
plt.title('Date')
plt.show()
```

```python
plt.figure(figsize=(8,8))
bees.time.value_counts().plot(kind = 'bar')
plt.title('Time')
plt.show()
```

```python
plt.figure(figsize=(6,6))
bees.location.value_counts().plot(kind = 'pie')
plt.title('Location')
plt.show()
```

```python
plt.figure(figsize=(6,6))
bees.subspecies.value_counts().plot(kind = 'pie')
plt.title('Subspecises')
plt.show()
```

```python
plt.figure(figsize=(6,6))
bees.health.value_counts().plot(kind = 'pie')
plt.title('Hive Health')
plt.show()
```

```python
plt.figure(figsize=(6,6))
bees.pollen_carrying.value_counts().plot(kind = 'bar')
plt.title('Pollen Carrying')
plt.show()
```

```python
plt.figure(figsize=(6,6))
bees.caste.value_counts().plot(kind = 'bar')
plt.title('Caste')
plt.show()
```
